```python
import tracery
```

```python
from tracery.modifiers import base_english
base_english
```

```python
rules = {
    "origin": "Utsav: You need some #object.capitalize##drugTerm.lowercase.s#.",
    "object": what_you_need,
    "drugTerm":poem
}

grammar = tracery.Grammar(rules)
grammar.add_modifiers(base_english)
for i in range(10):
    print("Me: I'm depressed.")
    print(grammar.flatten("#origin#"))
    print()
```
