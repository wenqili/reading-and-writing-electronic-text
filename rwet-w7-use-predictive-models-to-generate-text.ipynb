```python
import tracery
```

```python
from tracery.modifiers import base_english
base_english
```

```python
drug_stems_nodash = [item.replace("-","") for item in drug_stems_afterRNN if "-" in item ]
drug_stems_nodash
```

```python
rules = {
    "origin_1": "Me: I'm #my_feeling#.",
    "origin_2": "Utsav: Relax, you need some #object.capitalize##drugTerm.lowercase.s#.",
    "my_feeling": ["depressed"],
    "object": ["job", "visa", "sex", "food", "money", "career", "date"],
    "drugTerm":drug_stems_nodash
}

grammar = tracery.Grammar(rules)
grammar.add_modifiers(base_english)
print("Behind Tisch Building")
print("by: RNN(Recurrent Neural Network)")
print("")

for i in range(10):
    print(grammar.flatten("#origin_1#"))
    print(grammar.flatten("#origin_2#"))
    print()
```
