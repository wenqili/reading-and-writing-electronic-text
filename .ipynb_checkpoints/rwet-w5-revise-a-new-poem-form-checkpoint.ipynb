````python
ocupus = open("Ratatouille.txt").read().split()
````

remove empty lines in ocupus

````python
ocupus
````

````python
for i in range(0, len(ocupus)):
    ocupus[i] = ocupus[i].replace("?","")
    ocupus[i] = ocupus[i].replace(".","")
    ocupus[i] = ocupus[i].replace(",","")
    ocupus[i] = ocupus[i].replace("!","")
    ocupus[i] = ocupus[i].replace("-","")
````

````python
fo_words = [item for item in ocupus if item.startswith("fo")]
o_words = [item for item in ocupus if item.startswith("o")]
ran = [item for item in ocupus if len(item)>1]
ran = [item for item in ran if "fo" not in item]
````

````python
import tracery
from tracery.modifiers import base_english
````

````python
rules = {
    "origin": "#fo.capitalize# #ran.lowercase# #ran.lowercase# #ran.lowercase#, #o.capitalize#, #fo.capitalize# #ran.lowercase# #ran.lowercase# #ran.lowercase#.",
    "fo":fo_words,
    "o":o_words,
    "ran":ran
}
grammar = tracery.Grammar(rules)
grammar.add_modifiers(base_english)
raw = []
for i in range(0,10):
    print(grammar.flatten("#origin#"))
    raw.append(grammar.flatten("#origin#"))
````
